# Matplotlib Tutorial for Beginners

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure()
```

Figure = entire drawing

Axes = coordinate system that can display data

A figure can contain multiple axes objects next to one another

add_subplot(2, 2, 1): This method is used to add a subplot to the current figure. The parameters 2, 2, 1 indicate that you want a 2x2 grid of subplots, and you're currently selecting the first subplot (top-left corner). The numbers 2, 2 represent the number of rows and columns in the subplot grid, and 1 is the index of the current subplot.

```python
ax1 = fig.add_subplot(2, 2, 1)
```

```python
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

```python
data=np.random.randn(50,).cumsum()
```

```python
ax1.plot(data, 'k--')
```

```python
ax2.plot(data, "r-")
ax3.plot(data, "b.")
```

```python
fig
```

## Color, Markings, and Line Styles

'k-': This specifies the line style and color. 'k-' means a black solid line. You can customize this part based on your preferences (e.g., 'r--' for a red dashed line).
drawstyle='steps-post': This specifies that the line should be a stepwise line, connecting data points with horizontal and vertical lines.
label='Stepwise': This assigns a label to the line, which will be used in the legend.

```python
plt.plot(data,
         'k-',
         drawstyle='steps-post',
         label='Stepwise')
plt.legend()
```

```python
plt.plot(data, 'k-',
         drawstyle='steps-post',
         label='Stepwise')
plt.legend(loc='best')
```
