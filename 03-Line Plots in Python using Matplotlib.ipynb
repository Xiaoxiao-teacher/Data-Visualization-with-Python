# Line Plots in Python using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

By calling sns.set(), you apply the default Seaborn settings to your plots, including color palettes, grid styles, and other visual elements. It's a convenient way to make your visualizations more consistent and aesthetically pleasing.

```python
sns.set()
```

## Creating a Graphic Object

```python
x=np.linspace(0,10)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## Specifying Line Colors

```python
plt.plot(x, np.sin(x - 0), color='blue')
```

```python
plt.plot(x, np.sin(x - 1), color='g')
```

 You can also use a number between 0 and 1 to represent grayscale, where 0 is black, and 1 is white. The color='0.75' parameter sets the color of the line to a shade of gray (0.75 represents a light gray color).

```python
plt.plot(x, np.sin(x - 2), color='0.75')
```

Refer to https://matplotlib.org/stable/users/explain/colors/colors.html as needed

```python
plt.plot(x, np.sin(x - 3), color='#FFDD44')
```

## Specifying the Line Style

```python
plt.plot(x, x + 0, linestyle='solid')
```

```python
plt.plot(x, x + 1, linestyle='dashed')
```

```python
plt.plot(x, x + 2, linestyle='dashdot')
```

```python
plt.plot(x, x + 3, linestyle='dotted')
```

```python
plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':')
```

```python
plt.plot(x, x + 0, '-r')
```

```python
plt.plot(x, x + 1, '--b')
```

```python
plt.plot(x, x + 2, '-.g')
```

```python
plt.plot(x, x + 3, ':k')
```

## Specifying the Values of the Axes

```python
x
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-5, 20)
plt.ylim(-3, 3)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-5, 20)
plt.ylim(3,-3)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-5, 10, -3, 3])
```

the plt.axis('tight') command is used to automatically adjust the axis limits to tightly fit the data in the plot.

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

## Adding a Label to a Chart

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
```

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
plt.xlabel("x values")
plt.ylabel("sin(x) values")
```

plt.axis('equal') command is used to ensure that the aspect ratio of the plot is equal, meaning that the scaling of the x-axis and y-axis is the same.

```python
plt.plot(x, x+2, '-r', label='red')
plt.plot(x, x+4, ':b', label='blue')
plt.axis('equal')
plt.legend()
```

## Fill The Area

The alpha=0.30 parameter sets the transparency of the filled area to 30%, allowing you to see both the curve and the filled region.

```python
plt.fill_between(age, py_salary, alpha=0.30)
```

```python
ave_salary=57287
```

```python
plt.fill_between(age, py_salary, ave_salary, alpha=0.30)
```

## Graphic Styles

```python
plt.style.use('fivethirtyeight')
plt.plot(age,salary, "r-",marker=".")
plt.plot(age,py_salary, "b--",marker=".")
plt.legend(["All Developers","Python Developersr"])
```

The with plt.xkcd() block applies the xkcd sketch style temporarily for the specific plot commands inside it.

A plotting style inspired by the hand-drawn, comic-like visualizations often found in the webcomic xkcd

```python
with plt.xkcd():
    plt.plot(age,salary, "r-",marker=".")
    plt.plot(age,py_salary, "b--",marker=".")
    plt.legend(["All Developers","Python Developersr"])
```
