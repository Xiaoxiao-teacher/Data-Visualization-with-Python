# Bar Plot with Matplotlib in Python

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
```

```python
Math=np.random.randint(50,100,7)
Stat=np.random.randint(50,100,7)
CS=np.random.randint(50,100,7)
```

```python
classes=list("ABCDEFG")
classes
```

```python
plt.plot(classes, Math, label="Math Avg")
plt.plot(classes, Stat, label="Stat Avg")
plt.plot(classes, CS, label="CS Avg")
plt.legend()
plt.title("Average of Lessons")
plt.xlabel("Classes")
plt.ylabel("Averages")
```

```python
df=pd.DataFrame()
df['classes']=classes
df['Math']=Math
df['Stat']=Stat
df['CS']=CS
df=df.set_index('classes')
df.plot.bar()
```

## Bar Graph with Error Bars

```python
from sklearn.datasets import load_iris
```

iris.jpg

iris.data contains the feature data of the Iris dataset, which includes measurements of sepal length, sepal width, petal length, and petal width for different iris flowers. Each row represents a sample, and each column represents a feature.

```python
iris = load_iris()
iris.data
```

```python
mean = np.mean(iris.data, axis=0)
sd = np.std(iris.data, axis=0)
```

```python
interval = range(4)
```

The plt.barh function is used to create a horizontal bar plot. The xerr parameter is used to specify the error values. This is particularly useful when you have data points with associated uncertainties or errors, and you want to visually represent the variability or confidence intervals.

```python
plt.barh(interval ,mean, color="b",
         xerr=sd, alpha=0.7, align="center")
plt.title('Horizontal Bars')
```

```python
plt.barh(interval ,mean, color="b", xerr=sd,
         alpha=0.7, align="center")
plt.title('Horizontal Bars')
plt.yticks(interval, iris.feature_names)
```

```python
plt.bar(interval, mean, color="m",
        yerr=sd, alpha=0.7, align="center")
plt.title('Vertical Bars')
```

```python
plt.bar(interval, mean, color="m", yerr=sd,
        alpha=0.7, align="center")
plt.title('Vertical Bars')
plt.xticks(interval,
           iris.feature_names,
           rotation=70)
```

```python
#adjust the width of bars
plt.bar(interval, mean, color="m", yerr=sd,
        alpha=0.7, width=0.5, align="center")
plt.title('Vertical Bars')
plt.xticks(interval,
           iris.feature_names,
           rotation=70)
```
